```python
# write a function that can return the track record of 2 teams against each other
def head_to_head_record(team1, team2, ipl):
    """
    Returns the head-to-head track record of two teams in the IPL.

    Parameters:
    team1 (str): Name of the first team.
    team2 (str): Name of the second team.
    ipl_df (DataFrame): The IPL DataFrame containing match data.

    Returns:
    DataFrame: A DataFrame containing the results of matches between team1 and team2.
    """
    # Filter matches where Team1 and Team2 are the two teams
    mask1 = (ipl['Team1'] == team1) & (ipl['Team2'] == team2)
    mask2 = (ipl['Team1'] == team2) & (ipl['Team2'] == team1)

    # Combine the two conditions: it selects matches between team1 and team2 regardless of which team is Team1 or Team2.
    head_to_head_matches = ipl[mask1 | mask2]

    # Summarize results
    team1_wins = head_to_head_matches[head_to_head_matches['WinningTeam'] == team1].shape[0]
    team2_wins = head_to_head_matches[head_to_head_matches['WinningTeam'] == team2].shape[0]
    draws = head_to_head_matches[head_to_head_matches['WinningTeam'].isna()].shape[0]

    # Create a summary DataFrame
    summary_df = pd.DataFrame({
        'Team': [team1, team2],
        'Wins': [team1_wins, team2_wins],
        'Draws': [draws, draws]  # Assuming 'draws' counts as a tie/no result
    })

    return summary_df, head_to_head_matches[['Season', 'MatchNumber', 'Date', 'Venue', 'WinningTeam', 'Margin', 'Player_of_Match']]
# Example usage
summary, matches = head_to_head_record('Mumbai Indians', 'Chennai Super Kings', ipl)
print(summary)
print(matches)
```
